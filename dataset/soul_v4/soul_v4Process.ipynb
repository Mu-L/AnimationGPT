# fixNaN

```python
import numpy as np
from pathlib import Path

def check_files_nan(paths) -> bool:
    problems = []
    for i, p in enumerate(paths):
        t = np.load(p)
        if np.isnan(t).any():
            print(i, p)
            problems.append(p)
    return len(problems) > 0
```

```python
paths = [f for f in Path('soul_v4/new_joint_vecs/').iterdir() if f.is_file()]
check_files_nan(paths)
```

# 获取帧数写入&

```python
import pandas as pd
import numpy as np

df = pd.read_excel('v3_动画标注.一二简版.xlsx', sheet_name='Sheet1')
vecs_folder = 'soul_v4/new_joint_vecs'

for index, row in df.iterrows():
    animation_id = row['动画ID']
    vecs_file_path = os.path.join(vecs_folder, f'{animation_id}.npy')
    

    if os.path.exists(vecs_file_path):
        frames = np.load(vecs_file_path).shape[0]

        df.at[index, '标注.二简'] = row['标注.二简'].replace('&', str(frames))
        df.at[index, '标注.一简'] = row['标注.一简'].replace('&', str(frames))

df.to_excel('v4_动画标注.一二简版.xlsx', index=False)
```

# 删除多余的txt

```python
import os
import shutil

def find_and_move_files(txt_folder, npy_folder, destination_folder):
    txt_files = [os.path.splitext(file)[0] for file in os.listdir(txt_folder) if file.endswith(".txt")]

    npy_files = [os.path.splitext(file)[0] for file in os.listdir(npy_folder) if file.endswith(".npy")]

    unique_txt_files = set(txt_files) - set(npy_files)

    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    for file_name in unique_txt_files:
        txt_file_path = os.path.join(txt_folder, file_name + ".txt")
        destination_path = os.path.join(destination_folder, file_name + ".txt")

        shutil.move(txt_file_path, destination_path)
        print(f"Moved {file_name}.txt to {destination_folder}")


txt_folder_path = "texts"
npy_folder_path = "new_joints"
destination_folder_path = "delete"

find_and_move_files(txt_folder_path, npy_folder_path, destination_folder_path)
```

# 词性标注

```python
import spacy
import os
from tqdm import tqdm

nlp = spacy.load('en_core_web_sm')

def process_text(sentence):
    sentence = sentence.replace('-', '')
    doc = nlp(sentence)
    word_list = []
    pos_list = []
    for token in doc:
        word = token.text
        # if not word.isalpha():
        #     continue
        if not word.isalnum():
            continue
        if token.pos_ in ['NOUN', 'VERB'] and word.lower() != 'left':
            word_list.append(token.lemma_)
        else:
            word_list.append(word)
        pos_list.append(token.pos_)
    return word_list, pos_list

def annotate_line(line):
    caption = line.strip()
    start = 0.0
    end = 0.0

    word_list, pos_list = process_text(caption)
    tokens = ' '.join([f"{word}/{pos}" for word, pos in zip(word_list, pos_list)])
    
    return f"{caption}#{tokens}#{start}#{end}\n"

def annotate_txt(txt_folder):
    for txt_file in tqdm(os.listdir(txt_folder)):
        if txt_file.endswith(".txt"):
            file_path = os.path.join(txt_folder, txt_file)
            with open(file_path, 'r', encoding='utf-8') as f:
                lines = f.readlines()

            annotated_lines = [annotate_line(line) for line in lines]

            with open(file_path, 'w', encoding='utf-8') as f:
                f.writelines(annotated_lines)

if __name__ == "__main__":
    txt_folder = r'soul_v4\texts'
    annotate_txt(txt_folder)
```

# 检查存在中文字符的文本

```python
import os
import re
import shutil

def contains_chinese(text):
    """
    检查文本是否包含中文字符或中文标点，输出包含中文字符或中文标点的文件名称
    """
    pattern = re.compile('[\u4e00-\u9fa5]')
    return bool(pattern.search(text))

def find_files_with_chinese(src_folder):
    chinese_files = []
    error_files = []

    for filename in os.listdir(src_folder):
        if filename.endswith('.txt'):
            file_path = os.path.join(src_folder, filename)
            try:
                with open(file_path, 'r', encoding='utf-8') as file:
                    content = file.read()
                    if contains_chinese(content):
                        chinese_files.append(os.path.splitext(filename)[0])
            except Exception as e:
                error_files.append(os.path.splitext(filename)[0])
                print(f"Error reading file '{filename}': {str(e)}")

    return chinese_files, error_files

def copy_to_error_folder(file_list, src_folder, dest_folder):
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)

    for file_name in file_list:
        src_path = os.path.join(src_folder, file_name + '.txt')
        dest_path = os.path.join(dest_folder, file_name + '.txt')
        shutil.copy(src_path, dest_path)

source_folder = r's4'
chinese_files, error_files = find_files_with_chinese(source_folder)

fileName = 's4_errorTXT'
# 复制包含中文字符和标点的文件到errorTXT文件夹
copy_to_error_folder(chinese_files, source_folder, fileName)

# 复制导致程序报错的文件到errorTXT文件夹
copy_to_error_folder(error_files, source_folder, fileName)

print("Files with Chinese characters:")
for file_name in chinese_files:
    print(file_name)

print("\nError files:")
for file_name in error_files:
    print(file_name)
```
